# Dataloader

```python
class Load_data(Dataset):
    
    def __init__(self, file_path, transform=None):
        self.data = pd.read_csv(file_path)
        
    def __len__(self):
        return len(self.data)
    
    def __getitem__(self, index):
        X = self.data.iloc[index, 2:]
        label = self.data.iloc[index, 1]
        return torch.Tensor(X.values), label
```

```python
train_dataset = Load_data("final_data_train.csv", transform=None)
test_dataset = Load_data("final_data_test.csv", transform=None)
all_loader = Load_data("final_data.csv", transform=None)
```

```python
train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=8, shuffle=True)
all_loader = DataLoader(test_dataset, batch_size=8, shuffle=True)
```

# pyTorch Model = Best Best Best I feel at Rest Rest Rest 

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(348, 120)
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 2)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x

net = Net().cuda()
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(net.parameters(), lr=0.001, momentum=0.9)
```

```python
for epoch in range(10000):  
    running_loss = 0.0
    for i, data in enumerate(train_loader, 0):
        # get the inputs
        inputs, labels = data

        # zero the parameter gradients
        optimizer.zero_grad()

        # forward + backward + optimize
        outputs = net(inputs.cuda())
        loss = criterion(outputs, labels.cuda())
        loss.backward()
        optimizer.step()

        # print statistics
        running_loss += loss.item()
        if i % 2 == 0:    # print every 2000 mini-batches
            print('[%d, %5d] loss: %.3f' %
                  (epoch + 1, i + 1, running_loss / 2000))
            running_loss = 0.0

print('Finished Training')
```

```python
correct, wrong = 0, 0
for xt, yt in test_loader:
    y_pred = net(xt.cuda())
    for y1, y2 in zip(y_pred, yt):
        y1 = int(torch.argmax(y1))
        if y1 == y2:
            correct += 1
        else:
            wrong += 1
print(correct, wrong)            
```

```python
accuracy = 100 * correct / (wrong + correct)
print(f"Accuracy: {accuracy}%")
```

# Random Forest ew x2 (was implemented in the paper so... YOLO)

```python
paramGrid = ParameterGrid({
    'min_samples_leaf': [1,2,5,10, 20],
    'max_features': ['sqrt', 'log2', 0.4, 0.5, 0.6, 0.7,'auto'],
    'n_estimators': [10,20,80,100],
    'n_jobs': [-1],
    'random_state': [42],
    'max_depth':[2,3,5,10,20,30]
})

best_model, best_score, all_models, all_scores = pf.bestFit(RandomForestClassifier, paramGrid, 
                                                                                                                     X_train, y_train, X_test, y_test, 
                                                                                                                     metric=accuracy_score, scoreLabel='accuracy_score',
                                                                                                                     greater_is_better=True,verbose=False)
```

```python
best_model
```

```python
best_model.fit(X_train, y_train)
tst_s = accuracy_score(best_model.predict(X_test), y_test)
trn_s = accuracy_score(best_model.predict(X_train), y_train)
print(f"Test score: {tst_s} \n Train score: {trn_s}")
```
